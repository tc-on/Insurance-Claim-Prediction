```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
df= pd.read_csv('online_shoppers_intention.csv')
print(df.shape)
```

```python
df.isna().sum()
```

```python
df.head()
```

```python
sns.heatmap(df.corr())
```

```python
dft= df.iloc[:,:15]
```

```python
mon= dft['Month']
dft.drop({'Month'}, axis=1, inplace=True)
```

```python
col= dft.columns
```

```python
from sklearn.preprocessing import PowerTransformer
pt= PowerTransformer()
```

```python
dft = pt.fit_transform(dft)
```

```python
dft= pd.DataFrame(dft, columns=col)
```

```python
dft.head()
```

```python
dfr= df[['Month', 'VisitorType', 'Weekend', 'Revenue']]
dfr.head()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
for i in dfr.columns:
    dfr[i] = LabelEncoder().fit_transform(dfr[i])
```

```python
dfr.head()
```

```python
dft= dft.join(dfr)
```

```python
dft.head()
```

```python
# Lets try K means clustering first.
import scipy.cluster.hierarchy as sch
import sklearn
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
```

```python
rev= dft['Revenue']
dft.drop({'Revenue'}, axis=1, inplace=True)
```

```python
we= []
sc= []
for i in range(2, 15):
    km= KMeans(n_clusters=i, random_state=42)
    km.fit(dft)
    we.append(km.inertia_)
    lab= km.labels_
    sc.append(silhouette_score(dft,lab))
```

```python
plt.plot(range(2,15), we, 'r')
plt.xlabel('clusters')
plt.ylabel('SSD')
plt.show()
```

```python
# For KMeans optimum number of cluster is 3
```

```python
plt.plot(range(2,15), sc, 'b')
plt.xlabel('Sil score')
plt.ylabel('SSD')
plt.show()
```

```python
dft.shape
```

```python
km= KMeans(n_clusters=2, random_state=42)
km.fit(dft.iloc[:,:17])
```

```python
lab= km.labels_
dft['km_cl'] = lab
```

```python
# Agglomerative 

from sklearn.cluster import AgglomerativeClustering
import scipy.cluster.hierarchy as sch

dendrogram = sch.dendrogram(sch.linkage(dft, method='ward'))
```

```python
# For Agglo, optimum number of cluster is 2
```

```python
model = AgglomerativeClustering(n_clusters=2, affinity='euclidean', linkage='ward')
model.fit(dft)
labels = model.labels_
```

```python
dft['agg_cl'] = labels
dft['real'] = rev

pd.crosstab(dft['agg_cl'], dft['real'])
```

```python
# PCA
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
pca = PCA()
pca.fit(dft.iloc[:,:17])
```

```python
pca.explained_variance_
```
